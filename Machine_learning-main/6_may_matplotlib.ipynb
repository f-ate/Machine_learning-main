```python
import numpy as np
```

```python
level = np.array([1,2,3,4,5,6])
```

```python
salary = np.array([140,562,300,80,800,600])
```

```python
level
```

```python
salary
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(level,salary)
```

```python
plt.bar(level,salary,color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.bar(level,salary,color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing',fontsize=20)
plt.show()
```

```python
plt.scatter(level,salary,color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='red',marker='*',markersize=20,markerfacecolor='green',ls='--')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.pie(salary)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(level,salary)
plt.subplot(1,2,2)
plt.bar(level,salary)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.scatter(level,salary)
plt.subplot(1,3,2)
plt.bar(level,salary)
plt.subplot(1,3,3)
plt.plot(level,salary)
plt.tight_layout()
plt.show()
```

```python
plt.subplot(3,1,1)
plt.scatter(level,salary)
plt.subplot(3,1,2)
plt.bar(level,salary)
plt.subplot(3,1,3)
plt.plot(level,salary)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(level,salary)
axes.set_xlabel('Level')
axes.set_ylabel('Salary')
axes.set_title('Tesr')
plt.show()
# left,bottom,width,hieght
```

```python
incentives = np.array([10,20,30,5,40,25])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.4,0.4])
axes1.plot(level,salary)
axes2.plot(level,incentives)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(level,salary)
axes[1].plot(level,incentives)
plt.show()
```
