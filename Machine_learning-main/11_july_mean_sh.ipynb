```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import Axes3D
```

```python
centers = [[2,2,2],[7,7,7],[5,13,13]]
```

```python
X,_ = make_blobs(n_samples=150,centers=centers,cluster_std=0.60)
```

```python
X
```

```python
X.shape
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2])
plt.show()
```

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift()
```

```python
ms.fit(X)
```

```python
identified_cluster = ms.fit_predict(X)
```

```python
identified_cluster
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c=identified_cluster,cmap='rainbow')
plt.show()
```

```python
# Kmeans
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)
```

```python
kmeans.fit(X)
```

```python
identified_cluster_kmeans = kmeans.fit_predict(X)
```

```python
identified_cluster
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c=identified_cluster_kmeans,cmap='rainbow')
plt.show()
```

```python
wcss = []
for i in range(1,10):
    surbhi = KMeans(i)
    surbhi.fit(X)
    wcss_iter = surbhi.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

```python
number_cluster = range(1,10)
plt.plot(number_cluster,wcss)
plt.title('ELBow Method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```
