```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
test = pd.DataFrame(np.array([[1,345],[2,600],[3,200],[4,900],[5,100]]),columns=['Level','Salary'])
```

```python
test
```

```python
sns.barplot('Level','Salary',data=test)
plt.show()
```

```python
sns.scatterplot(x='Level',y='Salary',data=test)
```
