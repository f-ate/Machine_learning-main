```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.datasets import make_blobs
```

```python
centers = [[0.5,2],[-1,-1],[1.5,-1]]
```

```python
X, _ = make_blobs(n_samples=400,centers=centers,cluster_std=0.5,random_state=0)
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.4,min_samples=15)
```

```python
db.fit(X)
```

```python
identified_clusters = db.fit_predict(X)
```

```python
identified_clusters
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=identified_clusters,cmap='rainbow')
plt.show()
```
