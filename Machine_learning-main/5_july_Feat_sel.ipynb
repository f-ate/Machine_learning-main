```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
iris = pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
```

```python
iris['label'] = dataset.target
```

```python
iris
```

```python
X = iris.iloc[:,0:4].values
```

```python
X
```

```python
y = iris.iloc[:,4].values
```

```python
y
```

```python
X
```

```python
# removing low variance features
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.5)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_vt[:,0],X_vt[:,1],X_vt[:,2],c=y,cmap="rainbow")
plt.show()
```

```python
sel = VarianceThreshold(threshold=0.6)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
plt.scatter(X_vt[:,0],X_vt[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
# Select from model
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear')
```

```python
clf.fit(X,y)
```

```python
model = SelectFromModel(clf,prefit=True)
```

```python
X_xfs = model.transform(X)
```

```python
X_xfs
```

```python
plt.scatter(X_xfs[:,0],X_xfs[:,1],c=y,cmap='rainbow')
plt.show()
```
